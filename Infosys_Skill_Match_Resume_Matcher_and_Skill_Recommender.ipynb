## Matplotlib
python library for data visualization, primarily used to create static, animated, and interactive plots

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Line Plot

```python
x = np.arange(1,11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
plt.plot(x,y,color="g",linestyle=':',linewidth=2)
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# subplots
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.show()
```

### Bar Chart

```python
student = {'Bob':87,"Matt":56,"Sam":27}
names=list(student.keys())
marks=list(student.values())
```

```python
plt.bar(names,marks)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

```python
# horizontal bar plot
plt.barh(names,marks, color="g")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

## Seaborn
python library for creating attractive and informative statistical graphics, built on top of Matplotlib. It is used for data visualization to simplify creating plots like heatmaps, time series, and regression plots, making it easier to explore and understand data with a few lines of code.  

```python
import seaborn as sns
```

### Line Plot

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data =fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal",hue="event",style="event",markers=True,data=fmri)
plt.show()
```
